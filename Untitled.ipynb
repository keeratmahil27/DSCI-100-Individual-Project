```r
gender_and_subscribe <- players |>
ggplot(aes(x = gender, fill = subscribe)) +
    geom_bar(position = "dodge", color = "black") +
    labs( x = "Gender", y = "Count", fill = "Subscribed", title = "Gender vs. Subscription")
gender_and_subscribe

experience_and_subscribe <- players |>
ggplot(aes(x = experience, fill = subscribe)) +
    geom_bar(position = "dodge", color = "black") +
    labs( x = "Expeprience", y = "Count", fill = "Subscribed", title = "Experience vs. Subscription")
experience_and_subscribe

gender <- players |>
ggplot(aes(x = gender, fill = gender)) +
    geom_bar(color = "black") +
    labs( x = "Gender", y = "Count", fill = "Gender", title = "Gender of Players")
gender

experience <- players |>
ggplot(aes(x = experience, fill = experience)) +
    geom_bar(color = "black") +
    labs( x = "Experience", y = "Count", fill = "Experience", title = "Experience of Players")
experience
```

With these visulizations, I observe that the most subscribers are male and amateur players. However, the last two graphs show that majority of players are male and amateurs, so these graphs don't necessarily reflect how gender and experience impact subscription. For example, 100% of Agender players are subscribed to the newsletter, but there are only 2 Agender players, so this is not an accurate reflection of how likely Agender players are to subscribe.
